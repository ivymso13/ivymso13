# Pandas 소개
데이터를 분석하기 위해 주로 사용하는 파이썬 기반의 라이브러리로, 리스트에 비해 여러 기능이 추가되어 가장 많이 사용되는 라이브러리

## Series
1차원 데이터(열이 1개)를 저장하는 자료구조
- 인덱스(index)와 값(value)으로 구분된다.
- 별도의 인덱스를 지정하지 않으면 0번부터 시작된다.

### Series 선언 및 속성
pd.Series( 데이터, (인덱스, 이름) )<br>
선언 시, 인덱스와 이름은 생략 가능하며, 직접 데이터를 입력하거나 변수를 사용할 수 있음<br>
Ex. datas = pd.Series( [10,20,30], [1, 2, 3] )<br><br>
Series에서 주로 사용하는 속성은 아래 3가지로, Series 변수 뒤에 점(.)을 사용하여 표현한다. <br>
- .index : 각각의 데이터의 인덱스(키) 값
- .values : 모든 데이터의 값
- .dtype : 모든 데이터의 타입

```python
# Series 선언
import pandas as pd

cut_line = pd.Series([86.7, 87.3, 89.4]) # 값(value)만 입력
cut_line
```

```python
# Series 선언(2) >> 선언 시 인덱스에 2019, 2020, 2021 입력하기
cut_line = pd.Series([86.7, 87.3, 89.4], ) # 값(value)과 인덱스(index) 입력
print(cut_line)
```

```python
# Series 선언(3) >> 변수를 사용하여 선언하기
score = [86.7, 87.3, 89.4]
year = [2019, 2020, 2021]

cut_line = pd.Series() # 변수를 이용한 Series 선언
print(cut_line)
```

```python
# Series 선언(4) >> 딕셔너리 변수를 사용하여 선언하기
score = {2019:86.7, 2020:87.3, 2021:89.4}

cut_line = pd.Series( score )
print(cut_line)
```

```python
# Series 속성 사용하기
score = [86.7, 87.3, 89.4]
year = [2019, 2020, 2021]

cut_line = pd.Series(score, year)

print() # 인덱스 출력하기
print() # 값 출력하기
```

이번에는 전체 Series를 선택하지 말고, 안에 저장된 값을 사용하려면 어떻게 해야할까?<br>
여기서 우리는 인덱스(index)를 사용해서 특정 값을 따로 추출하여 사용할 수 있다. 만약 다음과 같이 선언된 cut_line 변수가 있다고 가정해보자.

> cut_line = pd.Series([86.7, 87.3, 89.4], [2019, 2020, 2021])

cut_line 변수는 Series로 선언되었으며, 총 3개의 값과 인덱스를 가지고 있다. 여기서 만약 인덱스 2020의 값을 사용하고 싶다면 우리는 <b>cut_line[2020]</b> 혹은 <b>cut_line.2020</b>으로 해당 값을 추출할 수 있다. 이렇게 인덱스의 값을 활용하면 우리가 원하는 데이터를 따로 추출할 수 있다.<br><br>
그럼 만약에 cut_line 변수가 다음과 같이 선언되었다면 <u>87.3</u>이라는 값을 출력하기 위해서는 어떻게 해야할까?

```python
cut_line = pd.Series([86.7, 87.3, 89.4])
print() # 87.3 출력하기
```

저장한 여러 개의 값 중에서 원하는 값을 사용하기 위해서는 인덱스를 정확하게 사용해야 한다. 특히 Series는 속성(열)이 1개이기 때문에 인덱스로 데이터를 구분하여 사용할 수 있다. 지금까지는 인덱스를 사용하여 단 1개의 데이터만 추출해보았다. <u><b>만약 2개 이상의 데이터를 추출하여 사용하고 싶다면</b></u> 어떻게 해야할까?<br><br>
여러 개의 데이터를 추출하고 싶다면 그만큼 인덱스를 기록해주면 된다. 원하는 데이터에 해당하는 인덱스를 모두 적으면 해당하는 값을 추출할 수 있다. 예를 들어, 이전의 cut_line 에서 [ 2019 ], [ 2020 ]에 해당하는 값을 추출하고 싶다면 <b>cut_line[ [2019, 2020] ]</b>과 같이 여러 개를 적어주면 된다. 여기서 중요한 점은 <u>리스트([ ])</u> 형태로 인덱스를 지정했다는 점으로, 인덱스를 여러 개 지정하기 위해서는 지정할 인덱스를 별도의 데이터 집합으로 지정할 필요가 있다.

```python
score = [86.7, 87.3, 89.4]
year = [2019, 2020, 2021]

cut_line = pd.Series(score, year)

print(cut_line[ [2019, 2020] ])
```

Series는 여러 개의 값을 하나의 변수에 저장할 수 있는 1차원적인 자료구조이다. Series은 리스트와 유사하지만 배열은 가지지 못한 속성이나 기능을 가지고 있기 때문에 데이터 분석에서 유용하게 사용된다. 예를 들어 모든 값의 평균을 구하고 싶다면 mean( )이라는 기능을 사용하면 손쉽게 구할 수 있다.

```python
cut_line = pd.Series([86.7, 87.3, 89.4])

cut_line.mean()
```

Series( )는 mean( ) 이외에도 여러 기능을 지원하고 있다. 그렇기에 특별한 기능이 필요하다면 검색을 통해 지원되는지 확인해보자. <br>
참고 ) https://pandas.pydata.org/docs/reference/api/pandas.Series.html

## DataFrame

- 인덱스(index)와 열(column), 그리고 행(row)로 구성
- 여러 개의 Series를 합친 자료구조

### DataFrame의 선언 및 속성
pd.DataFrame( 데이터, 인덱스, 열 )<br>
Series와 마찬가지로 인덱스와 열은 생략할 수 있음 <br>
Ex. data = pd.DataFrame( {'year':[2019, 2020, 2021], 'score':[92, 80, 88] } )
<br><br>
DataFrame의 주요 속성도 Series와 거의 유사하다.
- .index : DataFrame의 행(index)
- .columns : DataFrame의 열
- .values : DataFrame의 데이터
- .dtypes : DAtaFrame의 각각의 열의 데이터 타입

```python
# DataFrame 선언
df = pd.DataFrame( data=[ [2019, 2020, 2021], [92, 80, 88], ['김하나','김둘','김셋'] ], columns=['year', 'score', 'name'] )
df
```

```python
# DataFrame 선언(2)
df = pd.DataFrame( {'score': [92, 80, 88],
                    '1st_name': ['김하나','김둘','김셋']}, index=[2019, 2020,2021] )
df
```

```python
# DataFrame 속성 사용하기
df = pd.DataFrame( {'score': [92, 80, 88],
                    '1st_name': ['김하나','김둘','김셋']}, index=[2019, 2020,2021] )

print() # 열 출력하기
print() # 행 출력하기
print() # 값 출력하기
```

DataFrame은 이전에 학습한 Series와 거의 유사한 모습을 볼 수 있다. 다만, 이제는 하나의 속성(열)에 대해서 저장된 값이 많다는 점이 다른 점이라고 할 수 있다. 이렇게 행과 열로 구성된 데이터 집합에서 원하는 데이터를 추출하기 위해서는 열의 이름과 인덱스가 필요하다. 예를 들어, 다음과 같이 선언된 데이터가 있다고 가정하자.

> df = pd.DataFrame( { 'score': [92, 80, 88],'1st_name': ['김하나','김둘','김셋'] }, index=[2019, 2020,2021] )

여기서 만약에 2번째 이름인 '김둘'을 추출하고 싶다고 가정해보자. '김둘'은 현재 '1st_name'열에서 2020 인덱스 자리에 위치하고 있다. 이런 경우에는 <u>열->인덱스</u> 순서로 접근하면 '김둘'을 추출할 수 있다.


```python
df = pd.DataFrame( {'score': [92, 80, 88],
                    '1st_name': ['김하나','김둘','김셋']}, index=[2019, 2020,2021] )

print(df['1st_name'][2020])
```

DataFrame에서는 Series와는 다르게 항상 2개의 인덱스를 사용하게 된다. 사실 생각해보면 데이터가 표 형태로 저장되기 때문에 데이터에 접근하기 위해서는 행과 열 인덱스를 각각 지정해주어야 한다는 걸 알 수 있다. 하지만 인덱스를 2개 쓴다는 점 이외에는 Series와 동일하기 때문에 Series의 연장선이라고 생각하면 된다.<br><br>

이렇게 속성(열) 데이터는 이름을 사용하여 추출할 수 있다. 그렇다면 만약 행을 추출하고 싶다면 어떻게 할 수 있을까? 인덱스의 값을 추출하기 위해서는 별도의 기능인 <b>[ .loc[ ] ]</b>과 <b>[ .iloc[ ] ]</b>을 사용해야 한다. [ .loc[ ] ]과 [ .iloc[ ] ]은 사용하는 방법이 살짝 다르기 때문에 주의해서 사용해야 한다. 다만, tips 데이터에서는 두 함수의 사용법이 동일하기 때문에 아래 데이터를 사용해서 확인해보자.

> df = pd.DataFrame( { 'score': [92, 80, 88],'1st_name': ['김하나','김둘','김셋'] }, index=[2019, 2020, 2021] )

[ .loc[ ] ]은 인덱스의 <b>이름</b>을 사용하여 데이터를 추출하는 함수이다. 여기서 만약 df의 2020 인덱스의 값을 구하고 싶다면 df.loc[ 2020 ]으로 값을 추출할 수 있다.

```python
import pandas as pd
df = pd.DataFrame( { 'score': [92, 80, 88],'1st_name': ['김하나','김둘','김셋'] }, index=[2019, 2020,2021] )

df.loc[2020]
```

[ .iloc[ ] ]은 이름이 아닌 <b>순번</b>을 사용하여 추출하는 함수이다. 동일하게 2020 인덱스의 값을 추출하고 싶다면 df.iloc[ 2020 ]이 아닌, do.iloc[1]을 사용해야 한다. 이는 인덱스의 순번이 0부터 시작하여 2020의 값은 1번 인덱스 순번을 가지기 때문이다.

```python
df.iloc[1]
```

두 함수는 사실 행과 열 모두 추출할 수 있다. 함수 내부에 순서대로 행과 열의 범위를 적어주면 DataFrame에서 해당 범위의 데이터를 추출할 수 있다.

> 형식 : (i)loc[ [행], [열] ]

예를 들어, 행은 2019~2020까지, 열은 score를 추출하고 싶다면 우리는 다음과 같이 추출할 수 있다.

### (2-1) 관계형 시각화
관계형 시각화는 두 속성의 관계를 확인하기 위해 사용하는 시각화 방법이다. Tips에는 총 7개의 속성이 있으며, 이 중에서 연습으로 상관관계를 살펴볼 속성은 <b>total_bill 속성</b>과 <b>tip 속성</b>이다. 보통 가격이 비쌀수록 팁이 비싸진다는 통념이 있는데 실제 2개의 속성 사이의 상관관계가 있는지 살펴보기 위해서는 그래프를 그려보면 된다.

```python
tips = sb.load_dataset('tips')

# total_bill와 tip 사이의 상관관계 확인하기
sb.scatterplot(data=tips, x='total_bill', y='tip')
```

seaborn에는 그래프를 그릴 수 있는 여러 함수들이 있다. 그 중에서 경향성을 확인하기 가장 좋은 그래프는 산포도이다. 산포도들의 경향성을 보면 전체적으로 많은 금액을 결재할수록 팁을 많이 주는 걸 대략적으로 볼 수 있다. 이를 좀 더 확실하게 확인하기 위해서는 모든 값의 근사치를 직선 그래프로 그릴 수 있다. 여기에서는 <b>최소제곱법</b>을 사용하여 근사치 그래프를 그려보자.

최소제곱법은 모든 점 사이의 거리가 최소가 되는 직선 그래프를 그리는 방법이다. 직선 그래프( y = wx + b )를 그리기 위해서는 기울기 w와 y절편 b를 구해야 한다. w와 b는 다음 공식에 의해 확인할 수 있다. 

$\Large y = wx + b$일때,

- $\Large w = \Large\frac{\sum(x-\bar{x})(y-\bar{y})}{\sum(x-\bar{x})^{2}}$

- $\Large b = \Large\bar{y} - w\bar{x}$

w를 구한 다음, b를 구하면 원하는 결괏값을 얻을 수 있다.

```python
import numpy as np

# X와 Y의 평균 구하기
X = tips['total_bill']
meanX = tips['total_bill'].mean()

Y = tips['tip']
meanY = tips['tip'].mean()

w = ((X-meanX)*(Y-meanY)).sum() / ((X-meanX)**2).sum()
b = meanY - w*meanX

print(f"y = {w}x + {b}")
```

이렇게 구한 일차 방정식을 우리는 회귀식이라고 부른다. 회귀식을 산포도와 함께 그려보면 어떤 추세를 보이는지 확인할 수 있다.

```python
# 산포도 그리기
sb.scatterplot(data=tips, x='total_bill', y='tip', alpha=0.8)

# 회귀식 그리기
sb.lineplot(data=tips, x=X, y=(w*X+b), color='red')
```

산포도의 모습과 회귀선을 토대로 전체 가격(total_bill)과 팁(tip) 사이에는 유의미한 상관관계가 있음을 확인할 수 있다.

tips 전체 데이터가 아닌 tips_Sun 데이터만 사용하기에 scatterplot( )의 data 속성 값이 tips_Sun으로 달라진다는 점을 주목하자. 이렇게 추출한 데이터에도 회귀식을 구해서 그려볼 수 있다. 다만, 이렇게 여러 번 사용되는 알고리즘은 <b>함수</b>를 만들어 손쉽게 재사용할 수 있다. 다음과 같이 회귀선을 그려주는 함수를 만들어보자.

```python
def getRegression(df, xName, yName):
  # xName: x축 값
  X = df[xName]
  meanX = df[xName].mean()

  # yName: y축 값
  Y = df[yName]
  meanY = df[yName].mean()

  # 기울기와 절편 구하기
  w = ((X-meanX)*(Y-meanY)).sum() / ((X-meanX)**2).sum()
  b = meanY - w*meanX

  # y = w*X + b의 값 반환
  return [w, b]
```

이전 알고리즘과 살짝 달라진 점은 tips가 df로 바뀌었고, xName과 yName이 생겼다는 점이다. 어떤 DataFrame의 속성이든 회귀선을 구할 수 있게 일반화시켰다. 이렇게 일반화 시킨 함수를 사용하면 어떤 경우에서도 우리가 원하는 회귀선을 명령어 한 줄로 구할 수 있다는 장점이 있다. <br><br>

그럼 이전에 구해놓은 일요일 데이터의 회귀선을 구해보자. getRegression( )함수에 3가지 값을 넣으면 회귀선을 구할 수 있다. df에는 현재 분석할 DataFrame을, xName에는 x축 값의 이름을, yName에는 y축 값의 이름을 넣으면 된다.

```python
tips_Sun = tips[ tips['day'] == 'Sun' ]

# 산포도 그리기
sb.scatterplot(data=tips_Sun, x='total_bill', y='tip', color='black')

# 회귀선 그리기
regY = getRegression(tips_Sun, 'total_bill', 'tip')

x = tips_Sun['total_bill']
y = regY[0]*x + regY[1]
sb.lineplot(x=x, y=y, color='black')
```

위의 알고리즘을 활용하면 모든 요일마다 회귀선을 그릴 수 있다. 아래 그림을 참고하여 모든 요일마다 회귀선을 그려보자. 요일은 총 4개('Sun', 'Thur', 'Fri', 'Sat')가 있으며 각각 lineplot( )을 그려야 한다는 점을 참고해서 직접 그려보자.
