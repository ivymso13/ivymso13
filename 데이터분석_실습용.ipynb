# 주유소 데이터 분석하기
* 오피넷( https://www.opinet.co.kr/user/main/mainView.do ) 데이터로 분석하기

```python
# 1번 이동 시 메인 홈페이지가 나옴
driver.get('https://www.opinet.co.kr/searRgSelect.do')
driver.get('https://www.opinet.co.kr/searRgSelect.do')

# 경기도 선택
do = driver.find_element(By.ID, 'SIDO_NM0')
do.send_keys('경기')

time.sleep(1)

# 경기도의 모든 시/구/군 찾기
gu_list_raw = driver.find_element(By.XPATH, """//*[@id="SIGUNGU_NM0"]""")
gu_list = gu_list_raw.find_elements(By.TAG_NAME, 'option')

# option의 value 값 추출하기
gu_names = [ option.get_attribute('value') for option in gu_list ]
gu_names.remove('')
print(gu_names)
```

```python
# 시/군/구 바꾸기
select = driver.find_element(By.ID, 'SIGUNGU_NM0')
select.send_keys(gu_names[0]) # 가평군으로 변경

# 조회 버튼 누르기
btnPath = """//*[@id="searRgSelect"]"""
driver.find_element(By.XPATH, btnPath).click()

import time
time.sleep(2) # 잠시 대기

# 엑셀 다운로드
excelBtn = """//*[@id="glopopd_excel"]"""
driver.find_element(By.XPATH, excelBtn).click()
```

```python
# 경기도의 모든 시/군/구 파일 다운로드
from tqdm import tqdm

# 전처리
btnPath = """//*[@id="searRgSelect"]"""
excelBtn = """//*[@id="glopopd_excel"]"""

import os
for gu in tqdm(gu_names):
    # 시/군/구 선택
    element = driver.find_element(By.ID, 'SIGUNGU_NM0')
    element.send_keys(gu)
    time.sleep(2)
    
    # 조회버튼 클릭
    driver.find_element(By.XPATH, btnPath).click()
    time.sleep(1)
    
    # 엑셀 다운로드 클릭
    driver.find_element(By.XPATH, excelBtn).click()
    time.sleep(1)

    # 이름 변경하기
    os.rename('지역_위치별(주유소).xls', f'지역_위치별(주유소){gu}.xls')
    
```

```python
import pandas as pd
from glob import glob

# glob: 사용자가 제시한 조건에 맞는 파일명을 리스트 형식으로 반환
files = glob('지역*.xls')
files
```

```python
# xlrd 라이브러리 업데이트
!pip install --upgrade xlrd
```

```python
import pandas as pd

# 모든 파일 읽어오기
tmp = []
for file in files:
    t = pd.read_excel(file, header=2) # header 설정
    tmp.append(t)
  
모든 파일 합치기
station_raw = pd.concat(tmp)
station_raw.info()
```

```python
station_raw.head()
```

```python
# 휘발유와 경유 가격 비교하기
# 꼭 필요한 값만 가져오기
Data = pd.DataFrame({'Oil_store':station_raw['상호'],
                         '주소':station_raw['주소'],
                         '휘발유':station_raw['휘발유'],
                         '경유':station_raw['경유'],
                         '셀프':station_raw['셀프여부'],
                         '상표':station_raw['상표']})
Data.info()
```

```python
# 데이터 검증하기
# 시/군/구 추출하기
Data['시/군/구'] = [ address.split()[1] for address in Data['주소'] ]

Data['시/군/구'].unique() # 남양주, 광주, 여주군 발견
```

```python
# 오류 데이터 제거
def check(x):
  if x == '남양주' or x == '광주': return (x+'시')
  elif x == '여주군': return '여주시'
  else: return x

Data['시/군/구'] = Data['시/군/구'].apply(check)

Data['시/군/구'].unique()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt

# 휘발유 가격 비교하기
gasoline = pd.DataFrame({'지역':Data['시/군/구'], '가격':Data['휘발유']})
gasoline = gasoline[ gasoline['가격'] != '-' ]
gasoline = gasoline.astype({'가격':int})

gasoline.info()
```

```python
# 지역별 휘발유 가격 비교하기
plt.figure(figsize=(18,15))
plt.rc('font', family='NanumBarunGothic')

sb.barplot(x='가격', y='지역', data=gasoline)
```

```python
# 가격 순서대로 정렬하기
sort_gasoline = gasoline.groupby(['지역']).mean().sort_values('가격')

plt.rc('font', size=15)
plt.figure(figsize=(25, 10))
sb.barplot(x=sort_gasoline.index, y='가격', data=sort_gasoline, palette='RdPu')
```

```python
# 모든 지역의 평균 가격 확인하기
sort_gasoline = gasoline.groupby(['지역']).mean().sort_values('가격')
sort_gasoline
```

```python
# 바이올린 플롯을 활용하여 가격 분포 확인하기 > 안양시
gasoline_Anyang = Data[ (Data['시/군/구']=='안양시') & (Data['휘발유']!='-') ]
gasoline_Anyang = gasoline_Anyang.astype({'휘발유':int})

plt.figure(figsize=(15, 10))
plt.rc('font', size=15)
sb.violinplot(data=gasoline_Anyang, x='상표',y='휘발유')
```
